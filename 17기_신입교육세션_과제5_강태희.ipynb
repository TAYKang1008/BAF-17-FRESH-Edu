#### 8. 앙상블 학습(Ensemble Learning) 방법 중 Bagging과 Boosting의 차이점

각각 대표적인 알고리즘을 하나씩 제시해주세요.

병렬 학습인 Bagging은 각 모델은 독립적으로 훈련 데이터의 일부를 무작위로 샘플링하여 학습하고 이렇게 만들어진 여러 모델의 예측 결과를 평균(분류 문제에서는 다수결)하여 최종 예측을 생성한다. 순차 학습인 Boosting은 이전 모델의 예측 오류를 보완하도록 다음 모델을 학습시킨다. 즉, 이전 모델에서 잘 예측하지 못한 데이터에 더 큰 가중치를 부여하여 다음 모델이 해당 데이터에 집중하도록 한다.Bagging의 대표적인 알고리즘에는 Random Forest가 있다. Boosting의 대표적인 알고리즘에는 Gradient Boosting이 있다.

```python
# 연령에 따른 이탈률 나타내기

sns.boxplot(x='Churn', y='Age', data=df_train)
plt.title("연령 별 이탙률")
plt.xlabel('이탈률')
plt.ylabel('연령')

# 예측했던 가설과 달리 연령대가 높은 층에서 이탈률이 더 많음을 확인할 수 있음
```

```python
# 요금제에 따른 이탈률
# 둘 다 범주형 변수이기 때문에 bar plot 사용

sns.barplot(x='Subscription Type',y='Churn',data=df_train)

# 가설과 같이 basic에서 가장 이탈률이 큰 것을 확인할 수 있었으나 다른 요금제들과 크게 차이가 나지 않음을 확인할 수 있음
```

```python
# 요금제 유형별 이탈 여부 Count
sub_churn_counts = df_train.groupby(['Subscription Type', 'Churn']).size().unstack(fill_value=0)

# 각 계약 유형별 비율 계산
sub_churn_props = sub_churn_counts.div(sub_churn_counts.sum(axis=1), axis=0)

# 시각화
ax = sub_churn_props.plot(
    kind='bar',
    stacked=True,
    color=['lightcoral', 'gray'],
    figsize=(8,6)
)

# % Label 추가
for i, sub in enumerate(sub_churn_props.index):
    y_offset = 0
    for churn_status in sub_churn_props.columns:
        pct = sub_churn_props.loc[sub, churn_status] * 100
        if pct > 0:
            ax.text(
                i,
                y_offset + (sub_churn_props.loc[sub, churn_status] / 2),
                f"{pct:.1f}%",
                ha='center',
                va='center',
                fontsize=10
            )
        y_offset += sub_churn_props.loc[sub, churn_status]

# 그래프 꾸미기
plt.title("Churn Rate by Subscription Type", fontsize=14)
plt.ylabel("Proportion of Customers")
plt.xlabel("Subscription Type")
plt.ylim(0, 1)
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.PercentFormatter(1))
plt.legend(title="Churn", loc='best')
plt.tight_layout()
plt.show()
```

```python
# 계약기간에 따른 이탈률 

churn_rate_by_contract = pd.crosstab(df_train['Contract Length'], df_train['Churn'], normalize='index') * 100
churn_rate_by_contract[1].plot(kind='bar')
plt.title("Churn Rate by Contract Length")
plt.ylabel("Churn Rate (%)")

# Monthly 고객 대부분이 이탈한 것을 확인할 수 있음
```

```python
# 계약 유형별 이탈 여부 Count
contract_churn_counts = df_train.groupby(['Contract Length', 'Churn']).size().unstack(fill_value=0)

# 각 계약 유형별 비율 계산
contract_churn_props = contract_churn_counts.div(contract_churn_counts.sum(axis=1), axis=0)

# 시각화
ax = contract_churn_props.plot(
    kind='bar',
    stacked=True,
    color=['lightcoral', 'gray'],
    figsize=(8,6)
)

# % Label 추가
for i, contract in enumerate(contract_churn_props.index):
    y_offset = 0
    for churn_status in contract_churn_props.columns:
        pct = contract_churn_props.loc[contract, churn_status] * 100
        if pct > 0:
            ax.text(
                i,
                y_offset + (contract_churn_props.loc[contract, churn_status] / 2),
                f"{pct:.1f}%",
                ha='center',
                va='center',
                fontsize=10
            )
        y_offset += contract_churn_props.loc[contract, churn_status]

# 그래프 꾸미기
plt.title("Churn Rate by Contract Length Type", fontsize=14)
plt.ylabel("Proportion of Customers")
plt.xlabel("Contract Type")
plt.ylim(0, 1)
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.PercentFormatter(1))
plt.legend(title="Churn", loc='best')
plt.tight_layout()
plt.show()
```
