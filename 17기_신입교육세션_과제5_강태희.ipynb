#### 5. 분류 모델의 성능을 평가할 때 Precision과 Recall의 차이점은?

Precision이 중요한 상황과 Recall이 중요한 상황을 각각 적어주세요.

Precision은 모델이 양성이라고 예측한 것 중에서 실제로 양성인 비율을 나타낼 때 중요하고,   
Recall은 실제 정답 클래스에 속하는 인스턴스 중에서 모델이 얼마나 많은 부분을 정확하게 예측했는지를 나타내는 지표가 필요할 때 중요하다. 

#### 6. 결측치 처리 방법에는 여러 가지가 있습니다. 아래와 같은 상황에서는 어떤 결측치 처리 방법이 가장 적절한가요?

결측치가 40% 이상이면 삭제할 경우 정보가 부족해질 수 있고, 중위수로 대체할 경우 편향될 수 있다.  
그렇기에 다른 변수들을 사용해 결측값을 예측하는 방법을 사용해야 한다.

#### 8. 앙상블 학습(Ensemble Learning) 방법 중 Bagging과 Boosting의 차이점

각각 대표적인 알고리즘을 하나씩 제시해주세요.

병렬 학습인 Bagging은 각 모델은 독립적으로 훈련 데이터의 일부를 무작위로 샘플링하여 학습한다.  
순차 학습인 Boosting은 이전 모델의 예측 오류를 보완하도록 다음 모델을 학습시킨다. 즉, 이전 모델에서 잘 예측하지 못한 데이터에 더 큰 가중치를 부여하여 다음 모델이 해당 데이터에 집중하도록 한다.  
Bagging의 대표적인 알고리즘에는 Random Forest가 있다.  
Boosting의 대표적인 알고리즘에는 Gradient Boosting이 있다.  

#### 9. 다중공선성이란 무엇이며, 다중공선성이 심할 경우 해결 방법은 무엇인가요?



다중공선성은 회귀분석에서 독립변수들 간에 강한 상관관계가 나타나는 현상을 의미한다.

이런 경우 상관관계가 높은 변수들을 제거하거나 변수들을 적절하게 변환하여 상관관계를 낮추고 주성분 분석(PCA)과 같은 차원 축소 기법을 사용하여 변수 수를 줄인다. Ridge Regression 또는 Lasso Regression와 같은 규제 기법을 사용하여 다중공선성 문제를 해결할 수 있다.

### 가설

1. 연령과 이탈률의 관계: 나이가 적은 고객일수록 이탈률이 높을 것이다.  
   이유: 젊은 층은 요금제나 서비스를 자주 바꾸는 경향이 있을 수 있음.  

2. 요금제와 이탈률: Basic 요금제 고객이 Standard 요금제보다 이탈률이 높다.
   이유: Basic 요금제는 가격에 민감한 고객이 많고, 경쟁 서비스로 옮길 가능성이 높음.  

3. 계약 기간과 이탈률: 단기 계약(Monthly)이 장기 계약(Annual)보다 이탈률이 높다.  
   이유: 장기 계약은 중도 해지가 어렵고, 충성도 높은 고객이 많을 수 있음.  

5. 사용 빈도와 이탈률: 월별 사용 횟수(Usage Frequency)가 적을수록 이탈 확률이 높다.  
   이유: 자주 안 쓰는 서비스는 해지할 가능성이 높음.

6. 고객 지원 콜 횟수와 이탈률: 고객 지원 센터에 자주 전화한 고객일수록 이탈률이 높다.
   이유: 문제가 많거나 불만족했기 때문일 수 있음.  

7. 지불 지연(Payment Delay)과 이탈률: 지불 지연이 많은 고객일수록 이탈 확률이 높다.
   이유: 서비스에 대한 관심이 줄었거나 해지 의향이 클 수 있음.

8. 총 지출 금액과 이탈률: 총 지출 금액이 낮은 고객이 이탈할 가능성이 높다.
   이유: 서비스 이용률이 낮거나 가성비를 못 느꼈을 가능성

9. 마지막 상호작용 이후 일수와 이탈률: 마지막 접속 이후 시간이 길수록 이탈 확률이 높다.
   이유: 서비스에 관심이 줄었을 가능성

```python
# 요금제 유형별 이탈 여부 Count
sub_churn_counts = df_train.groupby(['Subscription Type', 'Churn']).size().unstack(fill_value=0)

# 각 계약 유형별 비율 계산
sub_churn_props = sub_churn_counts.div(sub_churn_counts.sum(axis=1), axis=0)

# 시각화
ax = sub_churn_props.plot(
    kind='bar',
    stacked=True,
    color=['lightcoral', 'gray'],
    figsize=(8,6)
)

# % Label 추가
for i, sub in enumerate(sub_churn_props.index):
    y_offset = 0
    for churn_status in sub_churn_props.columns:
        pct = sub_churn_props.loc[sub, churn_status] * 100
        if pct > 0:
            ax.text(
                i,
                y_offset + (sub_churn_props.loc[sub, churn_status] / 2),
                f"{pct:.1f}%",
                ha='center',
                va='center',
                fontsize=10
            )
        y_offset += sub_churn_props.loc[sub, churn_status]

# 그래프 꾸미기
plt.title("Churn Rate by Subscription Type", fontsize=14)
plt.ylabel("Proportion of Customers")
plt.xlabel("Subscription Type")
plt.ylim(0, 1)
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.PercentFormatter(1))
plt.legend(title="Churn", loc='best')
plt.tight_layout()
plt.show()
```

## 모델 학습

### 변수 스케일링

```python
# 학습 세트 만들기
# 총 지출 금액 대신 로그변환한 총 지출 금액을 넣을 것

X= df_train[['Age','Gender','Tenure','Usage Frequency','Support Calls','Payment Delay',
             'Subscription Type','Contract Length','Total Spend_log','Last Interaction']]
# X에는 ID, Churn 열을 제외

Y=df_train['Churn']
```

```python
# 모델 학습을 위해서는 문자형태로 되어있는 범주형 데이터를 숫자형으로 바꿔 주어야 한다. 
# pd.get_dummies() 함수는 명목변수*만 0과 1로 원-핫 인코딩을 해준다. 

X = pd.get_dummies(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, Y_train, Y_valid = train_test_split( X, Y, test_size=0.2, random_state=42)
```

## 모델 학습하기

- 이탈 고객을 예측하는 것이므로 분류를 사용할 것
- RandomForestClassifier, GradientBoostingClassifier, DecisionTreeClassifier으로 모델 적용해볼것

### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(random_state=42)
```

```python
rf.fit(X_train,Y_train)
```

```python
Y_predict_rf= rf.predict(X_valid)
Y_predict_rf
```

```python
from sklearn.metrics import accuracy_score

# 정확도 출력
print("정확도:", accuracy_score(Y_valid, Y_predict_rf))
```

### GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier

model_gb = GradientBoostingClassifier(max_depth=5, random_state=42)
model_gb.fit(X_train, Y_train)
```

```python
Y_predict_gb= model_gb.predict(X_valid)
```

```python
from sklearn.metrics import accuracy_score

# 정확도 출력
print("정확도:", accuracy_score(Y_valid, Y_predict_gb))
```

### DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier

dt= DecisionTreeClassifier(max_depth=5, random_state=42)
dt.fit(X_train, Y_train)
```

```python
Y_predict_dt= dt.predict(X_valid)
```

```python
from sklearn.metrics import accuracy_score

# 정확도 출력
print("정확도:", accuracy_score(Y_valid, Y_predict_dt))
```

GradientBoostingClassifier(max_depth=5, random_state=42)의 예측력이 제일 높음을 알 수 있음
